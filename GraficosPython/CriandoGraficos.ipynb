# Criando Gráficos em Python
O conjunto de dados Iris foi usado em R.A. O artigo clássico de Fisher de 1936, The Use of Multiple Measurements in Taxonomic Problems, e também pode ser encontrado no Repositório de Aprendizado de Máquina da UCI.

Inclui três espécies de íris com 50 amostras cada, bem como algumas propriedades sobre cada flor. Uma espécie de flor é linearmente separável das outras duas, mas as outras duas não são linearmente separáveis ​​uma da outra.

**Dataset:** https://www.kaggle.com/datasets/uciml/iris

# Gráficos
Aqui iremos criar alguns gráficos para estudo de Análise de Dados.

```python
# GRÁFICO DA COLUNA - 'SepalLengthCm'
# Determinando o tamanho da figura
plt.figure(figsize=(15, 8))
df_iris['SepalLengthCm'].plot(lw=3)   # Escolhendo a coluna para mostrar no gráfico
plt.title('Gráfico 01', size=15)
```

```python
# GRÁFICO DA COLUNA - 'SepalWidthCm'
# Gráfico de histograma
plt.figure(figsize=(15, 8))
df_iris['SepalWidthCm'].plot(kind='hist', color='green')
plt.title('Gráfico 02', size=15)
```

```python
# GRÁFICO DA COLUNA - 'PetalLengthCm'
# Gráfico de box
plt.figure(figsize=(15, 8))
df_iris['PetalLengthCm'].plot(kind='box')
plt.title('Gráfico 03', size=15)
```

```python
# GRÁFICO DA COLUNA - 'PetalWidthCm'
# Histograma com densidade
plt.figure(figsize=(10, 8))
df_iris['PetalWidthCm'].plot(kind='density', color='red', lw=3)   # lw --> Espessura da linha
df_iris['PetalWidthCm'].plot(kind='hist', color='blue', density=True)
plt.title('Gráfico 04', size=15)
```

# Gráfico de dispersão

```python
# Selecionando duas colunas para apresentação
# SepalLengthCm --> Coluna colocada no eixo X
# PetalLengthCm --> Coluna colocada no eixo Y
# scatter --> Gráfico de dispersão
df_iris.plot(x='SepalLengthCm', y='PetalLengthCm', kind='scatter', color='purple')
plt.title('Dispersão das flores', size=15)
plt.show()
```

```python
# Gráfico de compartimento hexágonal
df_iris = pd.DataFrame(np.random.randn(1000, 2), columns=["PetalWidthCm", "PetalLengthCm"])
df_iris["PetalLengthCm"] = df_iris["PetalLengthCm"] + np.arange(1000)
df_iris.plot.hexbin(x="PetalWidthCm", y="PetalLengthCm", gridsize=25);
```
